```python
import numpy as np
import pandas as pd
from pandas import Series
from pandas import DataFrame
from collections import defaultdict
from matplotlib import pyplot as plt

googlePlaystore = pd.read_excel("GooglePlaystore.xlsx")
googlePlaystore
```

[3 pts] Often there are outliers which do not match the overall data type. There is one record in this data where the "Reviews" has value "3.0M" which does not match the rest of the data. Remove that record.


```python
for i in range(0, len(googlePlaystore["Reviews"])):
    if googlePlaystore["Reviews"][i] == '3.0M':
        googlePlaystore = googlePlaystore.drop(i, axis = 0)
        break
googlePlaystore
```

[4 pts] Remove rows where any of the columns has the value "Varies with device".


```python
for j, i in googlePlaystore.iterrows():
    for k in i:
        if k == "Varies with device":
            googlePlaystore = googlePlaystore.drop(googlePlaystore.loc[googlePlaystore['App'] == i.App].index)
googlePlaystore
```

[5 pts] The values in the Android version column should be floats. Strip the trailing non-numeric characters from all values (ie. the words " and up"), so the result is a number. If there are multiple decimal places (eg. "x.y.z"), keep only the first two parts (eg "x.y"). For example, the value "4.1 and up" should be changed to "4.1". The value "4.5.6 and up" should be changed to "4.5". The value "5.6.7" should be changed to "5.6".

If there is a range (eg. 5.0 - 8.0), only consider the first number. For example, the value "5.0 - 8.0" should be changed to "5.0". The value "4.0.3 - 7.1.1" should be changed to "4.0".


```python
googlePlaystore["Android Ver"] = googlePlaystore["Android Ver"].str.extract(r'(\d+.\d+)', expand=False)
googlePlaystore
```

[5 pts] The "Installs" column must have integer values. For values that have commas, remove the commas. For values that have a '+' at the end, remove the '+'. Keep only those rows that have an integer value after these edits.


```python
googlePlaystore["Installs"] = googlePlaystore["Installs"].apply(lambda x : x.replace(',',''))
googlePlaystore["Installs"] = googlePlaystore["Installs"].apply(lambda x : x.replace('+',''))
googlePlaystore["Installs"] = googlePlaystore["Installs"].astype(int)
googlePlaystore
```

[5 pts] For missing rating values, if the number of reviews is less than 100 and installations is less than 50000, remove the row. Else, fill the missing value with the average value (rounded to 2 decimal places) for the Category of that row.


```python
s = dict(zip(googlePlaystore.groupby('Category').mean().index, googlePlaystore.groupby('Category').mean().Rating.round(2)))

for index, i in googlePlaystore.iterrows():
    
    if pd.isna(i.Rating):
        if i.Reviews < 100 and i.Installs < 50000:
            googlePlaystore = googlePlaystore.drop(googlePlaystore.loc[googlePlaystore['App'] == i.App].index)
        else:
            googlePlaystore.at[index,"Rating"] = s[googlePlaystore.at[index, 'Category']]
            
googlePlaystore
```

[6 pts] Preprocess the Size column to convert the "M" (millions) and "K" (thousands) values into integers. For instance, 8.7M should be converted to 8700000 and 2.4K should be converted to 2400.


```python
for index, i in googlePlaystore.iterrows():
    if 'k' in str(i.Size):
        replace = str(i.Size).replace('k', '')
        replace = int(float(replace) * 1000)
        googlePlaystore.at[index,"Size"] = replace
    if 'M' in str(i.Size):
        replace = str(i.Size).replace('M', '')
        replace = int(float(replace) * 1000000)
        googlePlaystore.at[index,"Size"] = replace
googlePlaystore
```

```python
stats = googlePlaystore.groupby("Category").describe().Rating
stats
```

[11 pts] Extract all "Free" apps from the master data frame. Then write a function that, given a numeric column e.g 'Rating'), will create and return a dataframe for the top 3 free applications in each category based on that column. Call the function on each of these columns:

a. Rating (gives top 3 most highly rated applications in each category)

b. Installs (gives top 3 most installed applications in each category)

c. Reviews (gives top 3 most reviewed applications in each category)

You don't need to do anything explicit to break ties.

Each of the returned dataframes have Category and App for the first two columns, and one of Rating (for a.), Installs (for b.), and Reviews (for c.) as the third column, as for instance:

```python
freeapps = googlePlaystore[googlePlaystore["Price"] == 0]
def top3(col):
    df = pd.concat([freeapps.Category, freeapps.App, col],axis=1)
    df = df.groupby(["Category"]).apply(lambda x: x.sort_values([col.name], ascending = False)).reset_index(drop=True)
    return df.groupby("Category").head(3).reset_index(drop=True)
    
top3RatedApps = top3(freeapps.Rating)
print(top3RatedApps)
top3InstalledApps = top3(freeapps.Installs)
print(top3InstalledApps)
top3ReviewedApps = top3(freeapps.Reviews)
print(top3ReviewedApps)
```

[4 pts] Find the average, maximum and minimum price of the paid applications.

```python
paidapps = googlePlaystore[googlePlaystore["Price"] != 0]
print(paidapps.Price.astype(float).describe())
print("\nTHE MEAN PRICE IS $" + str(paidapps.Price.astype(float).mean()))
print("THE MAXIMUM PRICE IS $" + str(paidapps.Price.astype(float).max()))
print("THE MINIMUM PRICE IS $" + str(paidapps.Price.astype(float).min()))
```

[9 pts] In the genre column, break the string of genres into a list. For example, ‘Art & Design; Creativity’ should be [‘Art & Design’, Creativity’].
Count the number of applications per genre and display it using a pie chart.

Hint: Read about DataFrame.explode()

```python
googlePlaystore["Genres"] = googlePlaystore["Genres"].str.split(";")
mapping = googlePlaystore["Genres"].explode().value_counts()

plt.pie(mapping.values,labels=mapping.index)
plt.show()
```

[7 pts] Display a box plot of ratings for "Business" and "Education" categories. The boxplots should be in the same plot.


```python
business = googlePlaystore[googlePlaystore["Category"] == "BUSINESS"]
education = googlePlaystore[googlePlaystore["Category"] == "EDUCATION"]
data = pd.DataFrame({"BUSINESS": business.Rating, "EDUCATION": education.Rating})
ax = data[["BUSINESS", "EDUCATION"]].plot(kind='box', title='Ratings for Business and Education')
ax.set_ylabel('ratings')
ax.set_xlabel('categories')
plt.show()
```
